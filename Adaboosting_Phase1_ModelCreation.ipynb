```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset.head()
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset.head()
```

```python
x=dataset.drop(['charges'],axis=1) # axis=0: Drop rows ; axis=1: Drop columns.
```

```python
x.head()
```

```python
y=dataset['charges']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
regressor=AdaBoostRegressor(random_state=96)
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
regressor.score(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
htregressor=AdaBoostRegressor(random_state=96,estimator=RandomForestRegressor(random_state=0),n_estimators=100,learning_rate=0.01)
```

```python
htregressor.fit(x_train,y_train)
```

```python
htregressor.score(x_test,y_test)
```

```python
htregressor.score(x_train,y_train)
```

```python
y_pred=htregressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r_score=r2_score(y_test,y_pred)
```
